```python
target_location = 'C:/Users/USJH706661/WSP O365/Bi-County Travel Demand Model Update - Documents/Task 06 2015 Model Calibration/Calibration Targets/Data/CTPP/'
ctpp_data = pd.read_csv(os.path.join(target_location,'A302100 - Total Workers (1) (Workers 16 years and over)_worker_flow.csv'), skiprows=2)
ctpp_data=ctpp_data[ctpp_data['Output']=='Estimate']
ctpp_data['RESIDENCE']=ctpp_data['RESIDENCE'].apply(lambda x: x.split(' County')[0])
ctpp_data['WORKPLACE']=ctpp_data['WORKPLACE'].apply(lambda x: x.split(' County')[0])
ctpp_data=ctpp_data.rename(columns={'Workers 16 and Over':'Workers'})
ctpp_data['Workers']=ctpp_data['Workers'].fillna('0')
ctpp_data['Workers']=ctpp_data['Workers'].astype(str)

ctpp_data['Workers']=ctpp_data['Workers'].apply(lambda x: int(x.replace(',','')))
ctpp_data[['RESIDENCE','WORKPLACE','Workers']].to_csv('../Calibration Targets/CTPP Workers CC Flow.csv', index=False)
```
