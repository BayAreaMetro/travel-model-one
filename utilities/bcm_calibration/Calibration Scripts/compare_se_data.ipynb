```python
import pandas as pd
import numpy as np

ctpp = pd.read_csv('../Calibration Targets/CTPP Workers CC Flow.csv')
bcm_sedata=pd.read_csv('../landuse/tazData.csv')
geo_crosswalk=pd.read_csv('../main/geo_crosswalk.csv')
county_dict=dict(zip(geo_crosswalk['TAZ'],geo_crosswalk['COUNTYNAME']))

bcm_sedata['WORKPLACE']=bcm_sedata['ZONE'].map(county_dict)
pd.options.display.max_columns=100
bcm_sedata.head()
```

```python
ctpp_total = ctpp.groupby('WORKPLACE').agg(Total_Workers_CTPP=('Workers','sum')).reset_index()
bcm_sedata_total = bcm_sedata.groupby('WORKPLACE').agg(Total_Workers_BCM=('TOTEMP','sum')).reset_index()

comparison=pd.merge(bcm_sedata_total, ctpp_total, on='WORKPLACE')

county_dict=['San Francisco',
             'San Mateo',
             'Santa Clara',
             'Alameda',
             'Contra Costa',
             'Solano',
             'Napa',
             'Sonoma',
             'Marin','San Joaquin']

county_sorter_index = dict(zip(county_dict, range(len(county_dict))))

comparison.sort_values(by='WORKPLACE',key=lambda x: x.map(county_sorter_index)).to_csv('../calibration_output/CTPP_SE_Comparison.csv', index=False)
```
