#### TAZ-Growth Geography(without PPA) and TAZ-TRA(within GGnonPPA) - for Travel Model Parking Pricing Strategy Input (DBP)
* The reason for not using Tract20_GG crosswalk and Tract20_GGTRA crosswalk: Travel Model in the Parking pricing strategy, need to calculate the share of each TAZ's area in GG and in GG+TRA, therefore requires direct spatial overlay calculation.

```python
# create interim crosswalk between TAZ and GGnonPPA: "M:\Application\PBA50Plus_Data_Processing\crosswalks\interim\TAZ1454_ggnonppaPBA50plus_crosswalk_DBP.csv"
%run TAZ_Census_otherGeographies_overlay_crosswalk.py "M:/Data/GIS layers/TM1_taz/bayarea_rtaz1454_rev1_WGS84.shp" "TAZ1454"  "TAZ1454" "M:/Application/PBA50Plus_Data_Processing/Draft_Blueprint_Growth_Geographies/QAQC/gg_nonPPA_pba50plus.shp" "ggnonppaPBA50plus" "gg_id" "M:/Application/PBA50Plus_Data_Processing/crosswalks/interim" --scenario DBP

# create interim crosswalk between TAZ and GGnonPPATRA: "M:\Application\PBA50Plus_Data_Processing\crosswalks\interim\TAZ1454_ggnonppatraPBA50plus_crosswalk_DBP.csv"
%run TAZ_Census_otherGeographies_overlay_crosswalk.py "M:/Data/GIS layers/TM1_taz/bayarea_rtaz1454_rev1_WGS84.shp" "TAZ1454"  "TAZ1454" "M:/Application/PBA50Plus_Data_Processing/Draft_Blueprint_Growth_Geographies/QAQC/gg_nonPPA_tra_pba50plus.shp" "ggnonppatraPBA50plus" "ggtra" "M:/Application/PBA50Plus_Data_Processing/crosswalks/interim" --scenario DBP
```

```python
## TAZ - Growth Geography
# load interim file
taz_ggPBA50plus = pd.read_csv(os.path.join(crosswalk_interim_dir, 'TAZ1454_ggnonppaPBA50plus_crosswalk_DBP.csv'))
print('{} rows, {} unique GEOID'.format(taz_ggPBA50plus.shape[0], taz_ggPBA50plus['TAZ1454'].nunique()))
display(taz_ggPBA50plus.head(3))

## TAZ - TRA (within GG)
# load interim file
taz_ggtraPBA50plus = pd.read_csv(os.path.join(crosswalk_interim_dir, 'TAZ1454_ggnonppatraPBA50plus_crosswalk_DBP.csv'))
print('{} rows, {} unique GEOID'.format(taz_ggtraPBA50plus.shape[0], taz_ggtraPBA50plus['TAZ1454'].nunique()))
display(taz_ggtraPBA50plus.head(3))

# create data needed for travel model Parking Pricing strategy

taz_gg_ggtraPBA50plus = pd.merge(
    taz_ggPBA50plus.rename(columns={'intersection_sq_m': 'area_within_GGnonPPA_sq_m',
                                    'area_share': 'pct_area_within_GGnonPPA'}),
    taz_ggtraPBA50plus.rename(columns={'intersection_sq_m': 'area_within_GGnonPPA_TRA_sq_m',
                                       'area_share': 'pct_area_within_GGnonPPA_TRA'}),
    on=['TAZ1454', 'base_sq_m', 'scen'],
    how='outer'
)
print('check merge: {} rows, {} unique TAZ id'.format(
    taz_gg_ggtraPBA50plus.shape[0], taz_gg_ggtraPBA50plus['TAZ1454'].nunique()
))
display(taz_gg_ggtraPBA50plus.head(3))

# fillna
for i in ['area_within_GGnonPPA_sq_m', 'pct_area_within_GGnonPPA', 'area_within_GGnonPPA_TRA_sq_m', 'pct_area_within_GGnonPPA_TRA']:
    taz_gg_ggtraPBA50plus[i].fillna(0, inplace=True)

# bring in 'area_sqmi' from PBA50 data
taz_gg_ggtraPBA50 = pd.read_excel(
    r'M:\Application\Model One\RTP2021\Blueprint\INPUT_DEVELOPMENT\parking_strategy\TAZ_intersect_GG_TRA.xlsx',
    sheet_name = 'bayarea_rtaz1454_rev1_WGS84')
display(taz_gg_ggtraPBA50.head(3))

taz_gg_ggtraPBA50plus = taz_gg_ggtraPBA50plus.merge(taz_gg_ggtraPBA50[['TAZ1454', 'area_sqmi']], on='TAZ1454', how='outer')
print('check merge: {} rows, {} unique TAZ id'.format(
    taz_gg_ggtraPBA50plus.shape[0], taz_gg_ggtraPBA50plus['TAZ1454'].nunique()
))

# write out the table for travel model

taz_gg_ggtraPBA50plus[['TAZ1454', 'area_sqmi', 'pct_area_within_GGnonPPA', 'pct_area_within_GGnonPPA_TRA', 'scen']].to_csv(
        os.path.join(tm_data_dir, 'parking_strategy', 'taz1454_GGnonPPA_TRA_crosswalk_DBP.csv'), index=False)

# write out additional QAQC data
taz_gg_ggtraPBA50plus['area_within_GGnonPPA_sqmi'] = taz_gg_ggtraPBA50plus['area_within_GGnonPPA_sq_m'] * SQMETER_TO_SQMILE
taz_gg_ggtraPBA50plus['area_within_GGnonPPA_TRA_sqmi'] = taz_gg_ggtraPBA50plus['area_within_GGnonPPA_TRA_sq_m'] * SQMETER_TO_SQMILE
taz_gg_ggtraPBA50plus['base_sqmi'] = taz_gg_ggtraPBA50plus['base_sq_m'] * SQMETER_TO_SQMILE
taz_gg_ggtraPBA50plus.to_csv(os.path.join(qaqc_dir, 'taz1454_GGnonPPA_GGnonPPATRA_crosswalk_DBP.csv'), index=False)
```
