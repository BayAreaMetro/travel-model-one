```python
folder = 'M:\\Application\\Model One\\RTP2021\\Blueprint\\INPUT_DEVELOPMENT\\Networks'
export_folder = os.path.join(folder, 'network_comparison', 'all_merged')
```

```python
plan_networks = {'2015Baseline':  'BlueprintNetworks_62\\net_2015_Baseline\\shapefile',
                 '2050Blueprint': 'BlueprintNetworks_64\\net_2050_Blueprint\\shapefile',
                 '2050Alt1':      'BlueprintNetworks_65\\net_2050_Alt1\\shapefile',
                 '2050Alt2':      'BlueprintNetworks_64\\net_2050_Alt2\\shapefile',
                 '2040Blueprint': 'BlueprintNetworks_64\\net_2040_Blueprint\\shapefile',
                 '2040Alt1':      'BlueprintNetworks_65\\net_2040_Alt1\\shapefile',
                 '2040Alt2':      'BlueprintNetworks_64\\net_2040_Alt2\\shapefile',
                 '2035Blueprint': 'BlueprintNetworks_64\\net_2035_Blueprint\\shapefile',
                 '2035Alt1':      'BlueprintNetworks_65\\net_2035_Alt1\\shapefile',
                 '2035Alt2':      'BlueprintNetworks_64\\net_2035_Alt2\\shapefile',
                 '2030Blueprint': 'BlueprintNetworks_64\\net_2030_Blueprint_TransitCuts_v01\\shapefile',
                 '2030Alt1':      'BlueprintNetworks_65\\net_2030_Alt1\\shapefile',
                 '2030Alt2':      'BlueprintNetworks_64\\net_2030_Alt2\\shapefile',
                 '2025Blueprint': 'BlueprintNetworks_64\\net_2025_Blueprint\\shapefile',
                 '2025Alt1':      'BlueprintNetworks_65\\net_2025_Alt1\\shapefile',
                 '2025Alt2':      'BlueprintNetworks_64\\net_2025_Alt2\\shapefile' 
                }

shapefiles = ['network_links.shp',
              'network_nodes.shp',
              'network_trn_lines.shp',
              'network_trn_links.shp',
              'network_trn_stops.shp']
```

```python
for shpfile in shapefiles:
    print(shpfile)
    merged_df = pd.DataFrame()
    for network_scen in plan_networks:
        print(network_scen)
        version_num = plan_networks[network_scen].split('\\')[0].split('_')[1]
        print('network version: {}'.format('v'+version_num))
        network_file = os.path.join(folder, plan_networks[network_scen], shpfile)
        network_df = gpd.read_file(network_file)
#         print('Read {} with dimension: {}'.format(network_file, network_df.shape))
        network_df['scenario'] = network_scen
        network_df['version'] = 'v'+version_num
        
        merged_df = pd.concat([merged_df, network_df])
#         print(merged_df.shape)
    
    print('Export {} rows of merge networks data'.format(merged_df.shape[0]))
#     merged_df.to_file(os.path.join(export_folder, shpfile))
```
