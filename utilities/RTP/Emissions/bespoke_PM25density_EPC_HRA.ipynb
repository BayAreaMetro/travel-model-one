```python
# get EPC shapes of both vintages, union the vintages, and dissolve by county

# get EPC shapes
epc_18_gdf = gpd.read_file(r'M:\Application\PBA50Plus_Data_Processing\EPC\equity_priority_communities_2025_acs2018_8750073566090738472.geojson')
epc_18_gdf = epc_18_gdf[['county_fip', 'epc_2050', 'geometry']]

epc_22_gdf = gpd.read_file(r'M:\Application\PBA50Plus_Data_Processing\EPC\draft_equity_priority_communities_pba2050plus_acs2022a_-5002738164289974036.geojson')
epc_22_gdf = epc_22_gdf[['county_fip', 'epc_2050p', 'geometry']]
epc_22_gdf['county_fip'] = epc_22_gdf['county_fip'].astype(str).str.zfill(3)

# Census county_fip to County name lookup
COUNTY_LOOKUP = pd.DataFrame({
    "county_fip": ["001", "013", "041", "055", "075", "081", "085", "095", "097"],
    "County": ["Alameda", "Contra Costa", "Marin", "Napa", "San Francisco", "San Mateo", "Santa Clara", "Solano", "Sonoma"]
})

```

```python
# set the CRS for both GeoDataFrames
epc_18_gdf = epc_18_gdf.to_crs(analysis_crs)
epc_22_gdf = epc_22_gdf.to_crs(analysis_crs)

epc_18_county = epc_18_gdf.dissolve(by=['county_fip'], as_index=False)
# display(epc_18_county)
epc_22_county = epc_22_gdf.dissolve(by=['county_fip'], as_index=False)
# display(epc_22_county)

epc_union_gdf = pd.DataFrame()

for county_id in ["001", "013", "041", "055", "075", "081", "085", "095", "097"]:
    epc_18_thisCounty = epc_18_county.loc[epc_18_county['county_fip'] == county_id][['geometry']]
    # epc_18_thisCounty.to_file(os.path.join(GIS_DIR, 'epc_18_thisCounty_{}.geojson'.format(county_id)), driver='GeoJSON')
    epc_22_thisCounty = epc_22_county.loc[epc_22_county['county_fip'] == county_id][['geometry']]
    # epc_22_thisCounty.to_file(os.path.join(GIS_DIR, 'epc_22_thisCounty_{}.geojson'.format(county_id)), driver='GeoJSON')
    epc_union_thisCounty = gpd.overlay(epc_18_thisCounty, epc_22_thisCounty, how='union')
    # epc_union_thisCounty.to_file(os.path.join(GIS_DIR, 'epc_union_thisCounty_{}.geojson'.format(county_id)), driver='GeoJSON')
    # print(f"County {county_id} union result:")
    # display(epc_union_thisCounty)
    epc_union_thisCounty_dissolved = epc_union_thisCounty.dissolve()
    epc_union_thisCounty_dissolved['county_fip'] = county_id
    epc_union_thisCounty_dissolved = epc_union_thisCounty_dissolved.merge(COUNTY_LOOKUP, on='county_fip', how='left')
    # display(epc_union_thisCounty_dissolved)
    # epc_union_thisCounty_dissolved.to_file(os.path.join(GIS_DIR, 'epc_union_thisCounty_dissolved_{}.geojson'.format(county_id)), driver='GeoJSON')
    epc_union_gdf = pd.concat([epc_union_gdf, epc_union_thisCounty_dissolved], ignore_index=True)

epc_union_gdf['COUNTY_EPC'] = epc_union_gdf['county_fip']+'_EPC'
display(epc_union_gdf)
print(type(epc_union_gdf))
```

```python
# get areas outside of EPC
non_EPC = gpd.overlay(counties_gpd[['fipco', 'County', 'geometry']], epc_union_gdf, how='difference')
# non_EPC.rename(columns={'NAME': 'County'}, inplace=True)
# use county id for field 'COUNTY_OBC' to be consistent with an subsequent script
# non_EPC['COUNTY_EPC'] = non_EPC['GEOID'].apply(lambda x: x[2:])
non_EPC.rename(columns={'fipco': 'COUNTY_EPC'}, inplace=True)
# display(non_EPC)

# put together
epc_all_gpd = pd.concat([epc_union_gdf, non_EPC], ignore_index=True)
# display(epc_all_gpd)

# calculate area in square miles
epc_all_gpd['area_mi2'] = epc_all_gpd.geometry.area * sqm_to_sqmi
print(f"Total area of EPC and non-EPC: {epc_all_gpd['area_mi2'].sum()} square miles")
print(epc_all_gpd.groupby('County')['area_mi2'].sum())

# write out the file for QAQC
epc_all_gpd_DIR = os.path.join(GIS_DIR, 'epc_all.shp')
epc_all_gpd.to_file(epc_all_gpd_DIR)
```
