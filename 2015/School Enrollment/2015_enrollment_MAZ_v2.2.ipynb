```python
# step 0: spatially join school data to MAZs using "contains" so schools on MAZ boundaries don't get dropped
```

```python
import numpy as np
import pandas as pd
```

```python
maz_school = pd.read_csv('./mazs_TM2_v2_2_schools_contains.csv')
enroll = pd.read_csv('./school_enrrollment_1516.csv')
```

```python
# check no duplicate schools exist
maz_school_notnull = maz_school[maz_school['JOIN_FID'] != -1]
print(len(maz_school_notnull))
maz_school_check = maz_school_notnull.drop_duplicates('JOIN_FID')
print(len(maz_school_check))
```

```python
# join schools and enrollment categories
maz_school_enroll = pd.merge(maz_school, enroll, left_on='cds_code', right_on='CDS_CODE', how='left')
maz_school_enroll.describe()
```

```python
# clean up columns and summarize enrollment by MAZ
maz_school_enroll = maz_school_enroll[['maz','ENR_TOTAL', 'ADULT']]
maz_school_enroll_sum = maz_school_enroll.groupby(['maz']).sum()
maz_school_enroll_sum.head()
```

```python
# check that enrollment totals are the same at the beginning and end
print(maz_school_enroll['ENR_TOTAL'].sum())
print(maz_school_enroll['ADULT'].sum())
print(maz_school_enroll_sum['ENR_TOTAL'].sum())
print(maz_school_enroll_sum['ADULT'].sum())
```

```python
# check that the correct no. of MAZs come out
len(maz_school_enroll_sum)
```

```python
maz_school_enroll_sum.to_csv('./2015_enrollment_MAZ_v2.2.csv')
```
